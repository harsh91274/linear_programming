## DEFINE THE SETS AND VARIABLES

```python
import pyomo.environ as pyo
from pyomo.opt import SolverFactory
model=pyo.ConcreteModel()
```

```python
# Define sets
model.SHIPMENTS = pyo.Set(initialize=shipments['shipment_id'].tolist())
model.LINEHAULS = pyo.Set(initialize=schedule.index.tolist())
```

```python
# Parameters
shipment_weight = dict(zip(shipments['shipment_id'], shipments['weight_kg']))
print(shipment_weight)

shipment_pallet = dict(zip(shipments['shipment_id'], shipments['pallet']))
print(shipment_pallet)

shipment_ready_time = dict(zip(shipments['shipment_id'], pd.to_datetime(shipments['ready_time_local'])))
print(shipment_ready_time)

shipment_due_time = dict(zip(shipments['shipment_id'], pd.to_datetime(shipments['due_time_local'])))
print(shipment_due_time)

#schedule_cost = dict(zip(schedule.index, schedule['rate_break'].apply(lambda x: x[0] if isinstance(x, list) else x)))
#print(schedule_cost)

schedule_travel_time = dict(zip(schedule.index, schedule['travel_hrs_with_timezone_offset']))
#print(schedule_travel_time)
```

```python
def parse_rate_break(rate_break):
    if isinstance(rate_break, str):
        rate_break = eval(rate_break)
    if isinstance(rate_break, list):
        rate_break = rate_break[0]
    return rate_break
```

```python
schedule['parsed_rate_break'] = schedule['rate_break'].apply(parse_rate_break)
schedule
```

```python
schedule['parsed_departure_time'] = schedule['departure_time'].apply(lambda x: datetime.strptime(x, "%H:%M").time())
```

```python
#PARAMETERS 

model.weight = pyo.Param(model.SHIPMENTS, initialize=shipment_weight)
model.pallet = pyo.Param(model.SHIPMENTS, initialize=shipment_pallet)


#model.ready_time = pyo.Param(model.SHIPMENTS, initialize=shipment_ready_time)
#model.due_time = pyo.Param(model.SHIPMENTS, initialize=shipment_due_time)
#model.cost = pyo.Param(model.LINEHAULS, initialize=schedule_cost)
#model.travel_time = pyo.Param(model.LINEHAULS, initialize=schedule_travel_time)

#model.crossdock_time = pyo.Param(initialize=4) 

model.ready_time = pyo.Param(model.SHIPMENTS, initialize={k: (v - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s') for k, v in shipment_ready_time.items()})
model.due_time = pyo.Param(model.SHIPMENTS, initialize={k: (v - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s') for k, v in shipment_due_time.items()})
model.travel_time = pyo.Param(model.LINEHAULS, initialize=dict(zip(schedule.index, schedule['travel_hrs_with_timezone_offset'])))
```

```python
model.x=pyo.Var(model.SHIPMENTS, model.LINEHAULS, domain=pyo.Binary)
x=model.x

#model.y = pyo.Var(model.SHIPMENTS, model.LINEHAULS, model.LINEHAULS, domain=pyo.Binary)
#y=model.y
```

## define the objective function and trade off parameter

```python
def calculate_ltl_cost(weight, rate_break):
    applicable_rate = max(rate_break.keys())
    for threshold, rate in sorted(rate_break.items()):
        if weight <= threshold:
            applicable_rate = rate
            break
    return weight * applicable_rate
```

```python
#trade off parameter
alpha = 0.5 #adjust for balance between cost and OTP
```

```python
def obj_rule(model):
    cost_term = sum(model.x[s, l] * (model.weight[s] / capacity.loc[0, 'max_weight_kgs']) * schedule['parsed_rate_break'][l] 
                    for s in model.SHIPMENTS for l in model.LINEHAULS if schedule.loc[l, 'rate_type'] == 'FTL') + \
                sum(model.x[s, l] * calculate_ltl_cost(model.weight[s], schedule['parsed_rate_break'][l])
                    for s in model.SHIPMENTS for l in model.LINEHAULS if schedule.loc[l, 'rate_type'] == 'LTL')
    
    otp_term = sum(model.x[s, l] * (model.due_time[s] - (model.ready_time[s] + model.travel_time[l] * 3600)) 
                   for s in model.SHIPMENTS for l in model.LINEHAULS if (model.due_time[s] - (model.ready_time[s] + model.travel_time[l] * 3600)) >= 0)
    
    return (1 - alpha) * cost_term - alpha * otp_term

model.obj = pyo.Objective(rule=obj_rule, sense=pyo.minimize)
```

## ADDING CONSTRAINTS TO THE MODEL

an assumption here is that one shipment can have a maximum of 3 routes used

```python
def route_constraint(model, s):
    return sum(model.x[s, l] for l in model.LINEHAULS) <= 1
model.route_con = pyo.Constraint(model.SHIPMENTS, rule=route_constraint)
```

```python
def weight_constraint(model):
    return sum(model.weight[s] * model.x[s, l] for s in model.SHIPMENTS for l in model.LINEHAULS) <= capacity.loc[0, 'max_weight_kgs']
model.weight_con = pyo.Constraint(rule=weight_constraint)
```

```python
def pallet_constraint(model):
    return sum(model.pallet[s] * model.x[s, l] for s in model.SHIPMENTS for l in model.LINEHAULS) <= capacity.loc[0, 'max_pallet_count']
model.pallet_con = pyo.Constraint(rule=pallet_constraint)
```

```python
# Output
output = []
for s in model.SHIPMENTS:
    for l in model.LINEHAULS:
        if model.x[s, l].value > 0:
            output.append({
                'shipment_id': s,
                'leg_id': l,
                'origin_region': shipments.loc[shipments['shipment_id'] == s, 'origin_region'].values[0],
                'destination_region': schedule.loc[l, 'destination_region'],
                'depart_time_local': schedule.loc[l, 'scheduled_depart_weekday_time_local'][0][1].strftime('%Y-%m-%d %H:%M:%S'),
                'drop_time_local': (schedule.loc[l, 'scheduled_depart_weekday_time_local'][0][1] + timedelta(hours=schedule_travel_time[l])).strftime('%Y-%m-%d %H:%M:%S'),
                'carrier_id': schedule.loc[l, 'carrier_id'],
                'truck_id': f"{schedule.loc[l, 'origin_region']}-{schedule.loc[l, 'destination_region']}-FTL-{schedule.loc[l, 'scheduled_depart_weekday_time_local'][0][1].strftime('%Y%m%d')}-0",
                'share_of_cost_based_on_weight': (model.weight[s] / capacity.loc[0, 'max_weight_kgs']) * schedule['parsed_rate_break'][l] if schedule.loc[l, 'rate_type'] == 'FTL' else calculate_ltl_cost(model.weight[s], schedule['parsed_rate_break'][l]),
                'met_otp': (schedule.loc[l, 'scheduled_depart_weekday_time_local'][0][1] + timedelta(hours=schedule_travel_time[l])) <= model.due_time[s]
            })

output_df = pd.DataFrame(output)
print(output_df)
```
