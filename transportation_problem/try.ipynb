```python
schedule = pd.read_excel("01-OR-RS-homework-data-template.xlsx", sheet_name='df_schedule')
```

```python
def convert_string_to_list(s):
    # Replace colons with quotes to make it a valid Python expression
    s = s.replace(", ", ",").replace("[","").replace("]","")
    return s

schedule['new_departure']=schedule['scheduled_depart_weekday_time_local'].apply(convert_string_to_list)
```

```python
schedule
```

```python
def parse_new_departure(s):
    tuples = s.split('),(')
    tuples = [t.replace('(', '').replace(')', '') for t in tuples]
    parsed_tuples = [tuple(t.split(',')) for t in tuples]
    return parsed_tuples
```

```python
schedule['parsed_departures'] = schedule['new_departure'].apply(parse_new_departure)
```
